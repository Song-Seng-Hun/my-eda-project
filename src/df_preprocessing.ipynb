```python
boardgame_price_new = pd.read_csv("../data/boardgame_price_df_new.csv", index_col=0)
boardgame_price_new = boardgame_price_new[~boardgame_price_new.duplicated()]
boardgame_price_new.head()
```

```python
boardgame_data_new = pd.merge(boardgame_data_new, boardgame_price_new, how='left', on='url')
boardgame_data_new = boardgame_data_new[~boardgame_data_new.duplicated()]
boardgame_data_new.head()
```

```python
boardgame_detail_data_new = pd.read_csv("../data/boardgame_detaildata_new.csv", index_col=0)
boardgame_detail_data_new = boardgame_detail_data_new[~boardgame_detail_data_new.duplicated()]
boardgame_detail_data_new.head()
```

```python
def striplist(strlist) :
    strlist = strlist.replace('Deck, Bag, and Pool Building', 'Deck Bag and Pool Building').replace('I Cut, You Choose', 'I Cut You Choose').replace('Worker Placement, Different Worker Types','Worker Placement Different Worker Types')
    strlist = strlist.replace('\"','').replace("\'","")
    return strlist.replace('[','').replace(']','')
```

```python
def year2int(str_year) :
    year = str_year.replace('(','').replace(')','')
    try :
        if int(year) > 1000 :
            return year
        else :
            return -1
    except :
        return -1
```

```python
def agerange(strage) :
    rangeage = strage.replace('\\u2013', '~').replace('Up to ', '0~').replace('All ages', '0~-1').replace('+', '~-1').replace('(no votes)', '-1~-1').replace('up', '~-1').replace(' and ','')
    if '~' not in rangeage :
        rangeage += '~' + rangeage
    return rangeage
```

```python
def isNotExpansion(kinds) :
    kindlist = str(str(kinds).split(','))
    if ('Expansion' in kindlist) or ('Game System' in kindlist) :
        return False
    else :
        return True
```

```python
boardgame_df = pd.concat([boardgame_data_new, boardgame_detail_data_new], axis = 1)
boardgame_df['출시일'] = boardgame_df['출시일'].apply(year2int).fillna(-1).astype('int').replace({-1: None})
boardgame_df['랭크'] = boardgame_df['랭크'].fillna(-1).astype('int').replace({-1: None})
boardgame_df['최소게임시간'] = boardgame_df['최소게임시간'].fillna(-1).astype('int').replace({-1: None})
boardgame_df['최대게임시간'] = boardgame_df['최대게임시간'].fillna(-1).astype('int').replace({-1: None})
boardgame_df['geek평점'] = boardgame_df['geek평점'].fillna(-1).astype('float').replace({-1: None})
boardgame_df['평균평점'] = boardgame_df['평균평점'].fillna(-1).astype('float').replace({-1: None})
boardgame_df['가격'] = boardgame_df['가격'].apply(lambda x : x.replace(',','')).fillna(-1).astype('float').replace({-1: None})
boardgame_df['진행방식'] = boardgame_df['진행방식'].astype('str').apply(striplist)
boardgame_df['종류'] = boardgame_df['종류'].astype('str').apply(striplist)
# boardgame_df.drop(labels='url', axis=1, inplace=True)
boardgame_df = boardgame_df[boardgame_df['종류'].apply(isNotExpansion)]
boardgame_df.iloc[30:40]
```

```python
boardgame_df.iloc[122:125]
```

```python
boardgame_df['적정나이'] = boardgame_df['적정나이'].apply(agerange)
boardgame_df['최소나이'] = boardgame_df['적정나이'].apply( lambda x : x.split('~')[0] )
boardgame_df['최대나이'] = boardgame_df['적정나이'].apply( lambda x : x.split('~')[1] )
boardgame_df.drop(labels='적정나이', axis=1, inplace=True)
boardgame_df.head()
```
